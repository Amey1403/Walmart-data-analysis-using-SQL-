```python
t['unit_price'] = t['unit_price'].replace('Nan', np.nan)  # Replace string 'Nan' with actual NaN
t['unit_price'] = t['unit_price'].str.replace('$', '', regex=False)  # Remove dollar sign
t['unit_price'] = pd.to_numeric(t['unit_price'],errors="coerce")  # Convert to float
```

```python
t['unit_price'] = t['unit_price'].fillna(100.0)
```
